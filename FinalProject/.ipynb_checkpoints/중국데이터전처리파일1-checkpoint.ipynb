```python
import pandas as pd
import numpy as np
import tqdm
```

```python
df = pd.read_csv("./전처리된 2015-2024 전체 데이터/2015~2024_중국기상데이터_전처리.csv")
```

```python
date = pd.date_range(start="2015-01-01", end="2024-05-20", freq = "3H")
```

```python
date = pd.DataFrame(date)[:-1]
```

```python
date.columns = ["일시"]
```

```python
# DataFrame을 79번 복제하여 axis=0 방향으로 추가
date_list = [date] * 80
result_df = pd.concat(date_list, axis=0).reset_index(drop=True)
```

```python
# datetime으로 변환 (머지를 위한 전처리)
df['일시'] = pd.to_datetime(df['일시'])

# datetime으로 변환 (머지를 위한 전처리)
result_df['일시'] = pd.to_datetime(result_df['일시'])

# 인덱스기준으로 머지 (left join)
merged_df = pd.merge(result_df, df, left_index=True, right_index=True, how="left")

# 결측치가 있는 행 확인
missing_dates = merged_df[merged_df.isnull().any(axis=1)]
```

```python
pdf = pd.pivot_table(merged_df,
                    index = "일시_x",
                    columns = "지점")
```

```python
# 수평구하는 함수
def horizontal_average(dataframe, datetime, column, threshold = 3, digits = 1):
    '''
    동일한 시간의 여러 지점 데이터를 확인한 후 해당 데이터들의 평균값을 반환하는 함수
    threshold 값을 입력하여 허용하는 NaN 값의 최대치를 설정할 수 있음
    '''  
    # 입력된 일시와 컬럼명에 해당하는 데이터 변수화
    data = dataframe.loc[datetime, column]
    
    # 임시로 nan값 입력
    mean = np.nan
    
    # data에 NaN값의 수가 threshold를 초과하는지 확인
    if data.isna().sum() <= threshold:
        # nan값을 제외한 값으로만 리스트 생성
        value_list = [i for i in data if not pd.isna(i)]
        
        # 평균값 산출 이후 반올림
        mean = np.mean(value_list)
        mean = round(mean, digits)
               
    return mean
```

```python
# 결측치를 평균값으로 치환하는 함수
def replace_nan_with_average(pdf, column_chunk_size=80, threshold=10, digits=1):
    # 각 행을 반복
    for idx, row in pdf.iterrows():
        # 각 행에서 80개씩 컬럼을 슬라이싱
        for i in tqdm(range(0, pdf.shape[1], column_chunk_size)):
            column_range = pdf.columns[i:i + column_chunk_size]
            mean_value = horizontal_average(pdf, idx, column_range, threshold, digits)
            
            # 슬라이싱한 컬럼 범위 내의 결측치를 평균값으로 치환
            pdf.loc[idx, column_range] = pdf.loc[idx, column_range].apply(lambda x: mean_value if pd.isna(x) else x)
```

```python
# 함수 호출
replace_nan_with_average(pdf, column_chunk_size=80, threshold=10, digits=1)
```
