```python
import pandas as pd
import numpy as np
from datetime import datetime
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("./기압만채우면됨.csv", index_col = 0)
```

```python
df["지점"] = df["지점"].astype(float)
```

```python
df.columns = ['일시', '지점', '지점명', '습도', '풍속', '풍향', '강수량', '현지기압', '해면기압', '변화량 기압',
       '이슬점 온도', '기온']
```

```python
pivot_df = df.pivot(index = "일시", columns = "지점", values = df.columns[3:])
```

```python
ss = StandardScaler()
```

```python
ss_pivot = ss.fit_transform(pivot_df)
```

```python
knn = KNNImputer(n_neighbors = 1)
knn_df = knn.fit_transform(ss_pivot)
```

```python
inverse_array = ss.inverse_transform(knn_df)
```

```python
inverse_df = pd.DataFrame(inverse_array)
```

```python
inverse_df.columns = pivot_df.columns
inverse_df.index = pivot_df.index
```

```python
inverse_df["풍향"] = inverse_df["풍향"].apply(lambda x: round(x, -1))
```

```python
inverse_df["풍향"] = inverse_df["풍향"].applymap(lambda x: 0 if x == 360.0 else x)
```

```python
dummy_df = pd.DataFrame(inverse_df.index).set_index("일시")

for loc in inverse_df["풍향"].columns:
    data = inverse_df.loc[:, ("풍향", loc)]
    dummy_data = pd.get_dummies(data)
    dummy_data.columns = dummy_data.columns.map(lambda x: str(float(loc)) + "_" + str(x))
    
    dummy_df = pd.concat([dummy_df, dummy_data], axis = 1)
```

```python
dummy_df.columns = dummy_df.columns.map(lambda x: ("풍향", x))
```

```python
inverse_df = inverse_df.drop("풍향", axis = 1)
```

```python
cn_weather = pd.concat([inverse_df, dummy_df], axis = 1)
```

```python
cn_weather.to_csv("./cn_weather.csv")
```
