### 풍향 단위 변경

- 현재 풍향은 0 ~ 36으로 표현되고 있음
    - 10을 곱하여 0 ~ 360도를 표현하도록 변경
    - 360의 경우 0으로 대체

```python
# 10배 곱하기 적용
mongol_df["16 방위 풍향"] = mongol_df["16 방위 풍향"].map(lambda x: 10 * x)
```

### 결측치 처리

- 습도, 강수량, 해면기압에서 다수의 음수값이 발견됨
    - 해면기압과 습도는 음수일 경우 결측치로 판단하여 interpolate linear로 대체
    - 강수량의 경우 결측치이거나 음수일 경우 0으로 대체

- 다른 데이터에서도 결측치가 일부 발견됨

```python
# 강수량 음수값일 때 값을 0으로 대체
mongol_df.loc[mongol_df["강수량"] < 0, "강수량"] = 0
```

```python
# 강수량의 결측치를 0으로 대체
mongol_df["강수량"] = mongol_df["강수량"].fillna(0)
```

```python
# 습도, 해면기압이 음수값일 경우 NaN으로 대체하여 구분하기 용이하게 만듦
mongol_df.loc[mongol_df["습도"] <= 0, "습도"] = np.nan
mongol_df.loc[mongol_df["해면기압"] <= 0, "해면기압"] = np.nan
```

```python
# 풍향 각도를 평균치로 맞추기 위한 함수
def average_angle_ignore_nan(degrees):
    """
    주어진 각도의 리스트에서 NaN 값을 무시하고 평균을 계산합니다.
    
    :param degrees: 각도의 리스트 (0-360도)
    :return: 평균 각도 (0-360도)
    """
    # NaN 값을 무시하고 유효한 각도만 선택
    valid_degrees = [deg for deg in degrees if not pd.isna(deg)]
    
    if not valid_degrees:
        return np.nan  # 유효한 각도가 없는 경우 NaN 반환
    
    # 유효한 각도를 라디안으로 변환
    radians = np.deg2rad(valid_degrees)
    
    # x, y 좌표 계산
    x_coords = np.cos(radians)
    y_coords = np.sin(radians)
    
    # x, y 좌표의 평균 계산
    x_mean = np.mean(x_coords)
    y_mean = np.mean(y_coords)
    
    # 평균 좌표를 각도로 변환
    mean_rad = np.arctan2(y_mean, x_mean)
    mean_deg = np.rad2deg(mean_rad)
    
    # 결과를 0-360도 사이의 값으로 변환
    mean_deg = mean_deg % 360
    
    return mean_deg
```

```python
mongol_df[mongol_df["16 방위 풍향"].isna()].index
```

```python
# 풍향이 결측치일 경우 전후 2개씩, 4개의 데이터 평균값으로 대체하기
for i in mongol_df[mongol_df["16 방위 풍향"].isna()].index:
    replace_value = round(average_angle_ignore_nan(mongol_df.loc[i-2:i+2, "16 방위 풍향"]), -1)
    mongol_df.loc[i, "16 방위 풍향"] = replace_value
```

```python
# 360 값을 0으로 대체
mongol_df.loc[mongol_df["16 방위 풍향"] == 360, "16 방위 풍향"] = 0
```

```python
# interpolate 함수로 결측치 대체
mongol_df = mongol_df.interpolate(method = "linear")
```

```python
# 시퀀스 데이터 생성 함수 변형버전
def create_sequences(data, time_steps=16):
    sequential = []
    for i in range(len(data) - time_steps):
        sequential.append(data.iloc[i:(i + time_steps)].values)
    return np.array(sequential)
```
