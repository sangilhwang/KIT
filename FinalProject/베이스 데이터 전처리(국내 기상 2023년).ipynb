```python
def kr_df_pre(data):
    global df_list # df_list를 전역변수로 사용
    global kr_2023_df # kr_2023_df를 전역변수로 사용
    
    df_list = []
    
    for i in data:
        # 데이터 파일 위치 변수명 생성
        filename = f"./kr{i}.csv"
        # 데이터 불러오기
        df = pd.read_csv(filename, encoding = "cp949")
        
        # 필요컬럼, 지역만 추출
        df = df.loc[:, ["지점", "지점명", "일시", "기온(°C)", "강수량(mm)", "풍속(m/s)", "풍향(16방위)",
                        "현지기압(hPa)", "해면기압(hPa)"]]
        df = df.loc[(df["지점명"] == "서울") | (df["지점명"] == "강화") | (df["지점명"] == "파주")
                    | (df["지점명"] == "동두천") | (df["지점명"] == "대전") | (df["지점명"] == "인천")
                    | (df["지점명"] == "수원")]
        
        # 컬럼명 변경
        df.columns = ["지점", "지점명", "일시", "16방위 풍향", "풍속", "강수량", "기온", "현지기압", "해면기압"]
        
        # 데이터 전처리
        df.drop("지점명", axis = 1, inplace = True) # 지점명 컬럼 제거
        df["month"] = df["일시"].map(lambda x : pd.to_datetime(x).month) # 일시컬럼에서 월 추출
        df["time"] = df["일시"].map(lambda x : pd.to_datetime(x).hour) # 일시컬럼에서 시간 추출
        df["일시"] = pd.to_datetime(df["일시"])
        df["강수량"].fillna(0, inplace = True) # 강수량 결측치 0으로 치환
        
        # 인덱스 재설정
        df.reset_index(drop = True, inplace = True)
        
#         # 풍속 전처리
#         for index_num in df.loc[df["풍속"].isna()].index:
#             df.iloc[index_num, 4] = round(df.loc[(index_num-30):(index_num+30),["풍속"]].mean(), 1)
            
#         # 기온 전처리
#         for index_num2 in df.loc[df["기온"].isna()].index:
#             df.iloc[index_num2, 6] = round(df.loc[(index_num2-30):(index_num2+30),["기온"]].mean(), 1)
          
        # 3시간 단위로 데이터 결합 및 전처리
        df["group"] = df.index // 3 # 3시간 단위로 동일한 값을 가지는 컬럼 추가
        df_group = pd.DataFrame() # 빈 데이터프레임 생성
        
        df_group["지점"] = df.groupby("group")["지점"].min()
        df_group["일시"] = df.groupby("group")["일시"].min() # 시간을 맞추기 위하여 최소값으로 처리
        df_group["16방위 풍향"] = df.groupby("group")["16방위 풍향"].median() # 방위이기때문에 중앙값으로 처리
        df_group["풍속"] = df.groupby("group")["풍속"].median() # 3시간 평균 풍속으로 처리
        df_group["강수량"] = df.groupby("group")["강수량"].sum() # 3시간동안 내린 강수량임으로 sum으로 합산
        df_group["기온"] = df.groupby("group")["기온"].median() # 3시간 평균 기온으로 처리
        df_group["현지기압"] = df.groupby("group")["현지기압"].median()
        df_group["해면기압"] = df.groupby("group")["해면기압"].median()
        df_group["month"] = df.groupby("group")["month"].min() # 월 그룹처리
        df_group["time"] = df.groupby("group")["time"].min() # 시간 처리
        
        # 데이터프레임 저장
        df_list.append(df_group)
        
    # 저장된 데이터프레임 결합
    kr_2023_df = pd.concat(df_list, ignore_index = True)
    
    return kr_2023_df
```
