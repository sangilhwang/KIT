```python
import pandas as pd
import numpy as np
import sqlite3
from datetime import datetime
```

```python
# 미세먼지 데이터 읽기
pm10_df = pd.read_csv("./Git/데이터/1. 종속변수/국내미세먼지농도.csv", index_col = 0, header = [0, 1])
pm25_df = pd.read_csv("./Git/데이터/1. 종속변수/국내초미세먼지농도.csv", index_col = 0, header = [0, 1])
```

```python
# 기상 데이터 읽기
kr_df = pd.read_csv("./독립변수/국내기상데이터.csv", index_col = 0, header = [0, 1])
```

- 서울 : 108
- 부산 : 159
- 대전 : 133
- 광주 : 156
- 강릉 : 105

```python
columns = ['지점번호', 'PM10', 'PM25', '강수량', '기온']
concat_df = pd.DataFrame(columns = columns)
```

```python
loc_code = ["105", "156", "133", "159", "108"]
```

```python
for i in range(0, 5):
    pm10 = pm10_df.iloc[11695:11719, i]
    pm25 = pm25_df.iloc[7:31, i]
    
    loc = loc_code[i]
    loc_ser = pd.Series(loc for _ in range(24))
    loc_ser.index = pd.date_range(start = "2019-01-02 00:00", end = "2019-01-04 21:00", freq = "3H").astype("str")
    
    weather = kr_df.xs(loc, axis = 1, level = 1).iloc[11695:11719, :2]
    
    df = pd.concat([loc_ser, pm10, pm25, weather], axis = 1)
    df.columns = columns = ['지점번호', 'PM10', 'PM25', '강수량', '기온']
    
    concat_df = pd.concat([concat_df, df], axis = 0)
```

```python
concat_df.index = concat_df.index.map(lambda x: datetime.strptime(x, "%Y-%m-%d %H:%M:%S"))
```

```python
concat_df["date"] = concat_df.index.map(lambda x: str(x)[:10])
```

```python
concat_df["time"] = concat_df.index.map(lambda x: str(x.hour))
```

```python
concat_df = concat_df.iloc[:, [5, 6, 0, 1, 2, 4, 3]]
```

```python
concat_df = concat_df.interpolate("linear")
```

```python
con = sqlite3.connect("./Git/web/db.sqlite3")
cursor = con.cursor()
```

```python
query = '''INSERT INTO predictions_predictions (date, time, loc, pm10, pm25, temperature, percipitation)
         VALUES (?, ?, ?, ?, ?, ?, ?)'''
```

```python
for i in range(len(concat_df)):
    data = concat_df.iloc[i, :].tolist()
    cursor.execute(query, data)
```
