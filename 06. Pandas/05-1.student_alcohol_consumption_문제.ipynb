# 문제3. 문자열 데이터를 입력 받고 capitalize() 메소드를 적용하는 capitalizer 함수 정의

```python
def capitalizer(string):
    return string.capitalize()   
```

# 문제4. Mjob과 Fjob열에 capitalizer 함수 적용된 결과 출력

```python
df_sac[["Mjob", "Fjob"]].applymap(capitalizer)
```

# 문제5. age가 17을 초과하는지 여부를 True, False로 표현하는 legal_drinker 열 생성

```python
def legal_drinker(age):
    if age > 17:
        return True
    
    else:
        return False
```

```python
df["legal_drinker"] = df["age"].map(legal_drinker)
df[["age", "legal_drinker"]].head()
```

```python
df_sac["age"].map(lambda x: True if x > 17 else False)
```

# 문제6. int타입의 데이터에만 10을 곱하기

```python
df_sac.applymap(lambda x: 10 * x if type(x) == int else x)
```

```python
def int_times_10(n):
    if type(n) == int:
        return 10 * n
    
    else:
        return n
```

```python
df = df.applymap(int_times_10)
df.head()
```
