```python
from tensorflow import keras
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
data = datasets.load_iris()
```

```python
data.data
```

```python
data.target
```

```python
# 모델 정의
def model_fn(a_layer = None):
    model = keras.Sequential()
    model.add(keras.layers.Dense(40, activation = "relu"))
    
    if a_layer:
        model.add(a_layer)
    model.add(keras.layers.Dense(3, activation = "softmax"))
    
    return model
```

```python
df = pd.DataFrame(data.data)
target = pd.Series(data.target)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df, target, test_size = 0.25, random_state = 7, stratify = target)
```

```python
# ss = StandardScaler()
# scaled_train = ss.fit_transform(x_train)
# scaled_test = ss.transform(x_test)
```

```python
scaled_train = x_train / df.max()
scaled_test = x_test / df.max()
```

```python
checkpoint_cb = keras.callbacks.ModelCheckpoint("./model/best-model.keras", save_best_only = True)
```

```python
model = model_fn(keras.layers.Dropout(0.3))
model.compile(optimizer = "adam", loss = "sparse_categorical_crossentropy", metrics = ["accuracy"])
history = model.fit(scaled_train, y_train, epochs = 500, validation_data = (scaled_test, y_test), callbacks = [checkpoint_cb])
```

```python
model.summary()
```

```python
plt.figure()

plt.plot(history.history["loss"], label = "loss")
plt.plot(history.history["accuracy"], label = "acc")
plt.plot(history.history["val_loss"], label = "val_loss")
plt.plot(history.history["val_accuracy"], label = "val_acc")

plt.legend(("train_loss", "train_acc", "val_loss", "val_acc"))
plt.xlabel("epoch")
plt.show()
```

```python
model.evaluate(scaled_test, y_test)
```
