### 데이터 제거해보기

```python
# 6까지의 숫자가 포함되지 않은 x_train2 / x_test2 생성
x_train2 = []
for i in x_train:
    arr_removed = [j for j in i if j not in remove_index[:16]]
    x_train2.append(arr_removed)
```

```python
x_test2 = []
for i in x_test:
    arr_removed = [j for j in i if j not in remove_index[:16]]
    x_test2.append(arr_removed)
```

```python
# 새로운 x_train2의 길이 측정
length2 = np.array([len(x) for x in x_train2])
```

```python
# 길이 평균값과 중간값
print(np.mean(length2), np.median(length2), np.min(length2), np.max(length2))
```

# 데이터 분할 및 간단한 전처리

```python
x_sub, x_val, y_sub, y_val = train_test_split(x_train2, y_train, test_size = 0.2, random_state = 7, stratify = y_train)
```

```python
maxlen = 20

sub_seq = pad_sequences(x_sub, maxlen = maxlen, truncating = "pre")
val_seq = pad_sequences(x_val, maxlen = maxlen, truncating = "pre")
test_seq = pad_sequences(x_test2, maxlen = maxlen, truncating = "pre")
```

```python
y_oh_sub = keras.utils.to_categorical(y_sub)
y_oh_val = keras.utils.to_categorical(y_val)
y_oh_test = keras.utils.to_categorical(y_test)
```

# 모델 구성

```python
model = keras.Sequential([
    keras.layers.Embedding(num_words, 128, input_shape = (maxlen,)),
    keras.layers.Dropout(0.3),
    keras.layers.Conv1D(128, 5, activation = "relu"),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPool1D(4),
    keras.layers.LSTM(64, dropout = 0.3, return_sequences = True),
    keras.layers.LSTM(64, dropout = 0.3),
    keras.layers.Dense(2, activation = "softmax")
])
```

```python
model.summary()
```

```python
rmsprop = keras.optimizers.RMSprop(learning_rate = 1e-4)
adam = keras.optimizers.Adam(learning_rate = 1e-4)
nadam = keras.optimizers.Nadam(learning_rate = 1e-4)
model.compile(loss = "categorical_crossentropy", optimizer = rmsprop, metrics = ["accuracy"])
early_stopping_cb = keras.callbacks.EarlyStopping(patience = 4, restore_best_weights = True)
checkpoint_cb = keras.callbacks.ModelCheckpoint("./model/model-nsmc-model.keras", save_best_only = True)
```

```python
history = model.fit(sub_seq, y_oh_sub, batch_size = 128, epochs = 100,
                   validation_data = (val_seq, y_oh_val),
                   callbacks = [checkpoint_cb, early_stopping_cb])
```

```python
plt.figure()

plt.plot(history.history["loss"], label = "loss")
plt.plot(history.history["accuracy"], label = "acc")
plt.plot(history.history["val_loss"], label = "val_loss")
plt.plot(history.history["val_accuracy"], label = "val_acc")

plt.legend(("train_loss", "train_acc", "val_loss", "val_acc"))
plt.xlabel("epoch")
plt.show()
```

```python
model.evaluate(test_seq, y_oh_test)
```
