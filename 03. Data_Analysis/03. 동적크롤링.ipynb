# selenium으로 DOM에 접근하는 방법

- 단일 객체 반환(bs4의 find()/select_one() 와 같은 형태)
    - find_element

- 리스트 객체 반환(bs4의 find_all()/select() 와 같은 형태)
    - find_elemnents

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
from urllib.request import urlopen
from selenium.webdriver.common.keys import Keys
```

```python
# 웹 접속하기
url = "https://www.naver.com"

driver = webdriver.Chrome()

driver.get(url) # url 접속
```

## 원하는 태그를 id로 찾는 방법

```python
url = "https://pjt3591oo.github.io"

driver = webdriver.Chrome()

driver.get(url)
```

```python
select_id = driver.find_element(by = By.ID, value = "nav-trigger")
```

```python
print(select_id)
```

```python
print(select_id.tag_name)
```

```python
print(select_id.text)
```

## 원하는 태그를 tag_name으로 찾는 방법

```python
select_tag_p = driver.find_element(by = By.TAG_NAME, value = "p")

print(select_tag_p)
print(select_tag_p.tag_name)
print(select_tag_p.text)
```

```python
select_tag_p = driver.find_elements(by = By.TAG_NAME, value = "p")

print(select_tag_p)
```

## 원하는 태그를 class_name으로 찾는 방법

```python
select_class = driver.find_element(by = By.CLASS_NAME, value = "p")

print(select_class)
print(select_class.tag_name)
print(select_class.text)
```

```python
select_class = driver.find_elements(by = By.CLASS_NAME, value = "p")

print(select_class)
```

## 원하는 태그를 css_selector로 찾는 방법

- bs4의 select()와 동일

```python
select_css = driver.find_element(by = By.CSS_SELECTOR, value = "div.p")
print(select_css)
print(select_css.tag_name)
print(select_css.text)
```

```python
select_css = driver.find_elements(by = By.CSS_SELECTOR, value = "div.p")
print(select_css)
```

## selenium 활용

```python
url = "https:/pjt3591oo.github.io/search"

driver = webdriver.Chrome()

driver.get(url)

search_box = driver.find_element(By.CSS_SELECTOR, "input#search-box")

search_box.click()

search_box.send_keys("db")
# 엔터키 - Keys 에는 키보드의 모든 특수키가 있음
search_box.send_keys(Keys.ENTER)

soup = BeautifulSoup(driver.page_source, "lxml")

items = soup.select("ul#search-results > li")

for i in items:
    print(i.select_one("a > h3").string)
    print(i.select_one("p").string)
```

# 예제2. 네이버 지식백과

- feat 묵시적 대기
- 코드와 코드 사이에 페이지로드를 위해 대기해야 하는 시간이 필요한 경우

```python
url = "https://www.naver.com/"

driver = webdriver.Chrome()
driver.get(url) # 페이지 이동

driver.implicitly_wait(3) # 묵시적 대기, 페이지가 활성화되기를 최대 3초까지 기다린다

search = driver.find_element(By.CSS_SELECTOR, "input#query")

search.send_keys("고슴도치")

search.send_keys(Keys.ENTER)

driver.implicitly_wait(3)

# 요소 찾기
post = driver.find_element(By.CSS_SELECTOR, "a.area_text_title")
post.click()
```

# 예제 3. 스타벅스

```python
import time
```

```python
url = "https://www.starbucks.co.kr/store/store_map.do?disp=locale"

driver = webdriver.Chrome()
driver.get(url)

driver.implicitly_wait(3)

city = driver.find_elements(By.CSS_SELECTOR, "a.set_sido_cd_btn")[0]
city.click()

driver.implicitly_wait(3)

area = driver.find_elements(By.CSS_SELECTOR, "a.set_gugun_cd_btn")[13]

area.click()

time.sleep(3)

soup = BeautifulSoup(driver.page_source, "lxml")

for i in soup.select("ul.quickSearchResultBoxSidoGugun > li > strong"):
    print(i.string.strip())
```
