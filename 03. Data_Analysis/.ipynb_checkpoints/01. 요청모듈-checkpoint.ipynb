```python
# 없는 url 주소에 요청을 보내면
# 다음은 없는 링크 요청

url = "https://www.naver.com/a"

res = rq.get(url)

# 응답코드 출력하기
print(res.status_code)
```

```python
# 응답코드를 활용한 조건분기
def url_check(url):
    res = rq.get(url)
    
    print(res)
    
    sc = res.status_code
    
    if sc == 200:
        print(f"{url} 요청 성공")
        
    elif sc == 404:
        print(f"{url} 찾을 수 없음")
        
    else:
        print(f"{url} 알 수 없는 에러 : {sc}")
        
```

```python
url_check("https://www.naver.com/")
```

```python
url_check("https://www.naver.com/a")
```

```python
# 헤더 가져오기
url = "https://www.naver.com/"

res = rq.get(url)

print(res)
print(res.headers)
```

```python
# html 코드 보기
print(res.text)
```

```python
# 바이너리 형태로 값을 가져오기 때문에 res.text에서 한글이 깨지는 현상을 방지할 수 있음
print(res.content)
```

```python
# 인코딩 확인
res.encoding
```

```python
# 쿼리스트링 데이터 생성 / 유지보수, 가독성 면에서 유리함
res = rq.get(url, params = {"key1" : "value1", "key2" : "value2"})

print(res.url)
```

```python
rq.get("https://www.naver.com/?key1=value1&key2=value2")
```

# urllib 으로 웹페이지 접속

```python
from urllib.request import urlopen
```

```python
url = "https://www.naver.com"

page = urlopen(url) # 요청 객체를 이용해 요청

print(page)
```

```python
print(page.code)
print("")
print(page.headers)
print("")
print(page.url)
print("")
print(page.info().get_content_charset())
```

```python
print(page.read())
```
